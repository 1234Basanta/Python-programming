The following function will eliminate the comma only however, the data is still an object not an integer.

```python
df = df.apply(lambda x: x.str.replace(',', ''))
df
```

As you can see the function ".astype(int) will replace the data object into integers. 

```python
df['Population'] = df['Population'].astype(int)
df['Deaths'] = df['Deaths'].astype(int)
df.info()
```
